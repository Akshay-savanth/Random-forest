```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
fraud=pd.read_csv("Fraud_check.csv")
fraud
```

```python
label_encoder=preprocessing.LabelEncoder()
fraud["Undergrad"]=label_encoder.fit_transform(fraud["Undergrad"])
fraud["MaritalStatus"]=label_encoder.fit_transform(fraud["MaritalStatus"])
fraud["Urban"]=label_encoder.fit_transform(fraud["Urban"])
```

```python
daf=pd.DataFrame(fraud.TaxableIncome)
daf.loc[fraud["TaxableIncome"]<=30000,'Risky or Not']='NOT RISKY'
daf.loc[fraud["TaxableIncome"]>30000,'Risky or Not']='Risky'
z=daf
z
```

```python
fraud.insert(6,"Risk_status",z['Risky or Not'])
fraud
```

```python
label_encoder=preprocessing.LabelEncoder()
fraud["Risk_status"]=label_encoder.fit_transform(fraud['Risk_status'])
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
array=fraud.values
X=array[:,[0,1,3,4,5]]
Y=array[:,6]
```

```python
seed=8
test_size=0.35
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=test_size,random_state=seed)
```

```python
model=XGBClassifier()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
predictions=[round(value)for value in y_pred]
```

```python
accuracy=accuracy_score(y_test,predictions)
print("Accracy:%.2f%%" % (accuracy * 100.0))
```
